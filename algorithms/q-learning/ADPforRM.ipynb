```python
import pandas as pd

df = pd.read_csv("./US Airline Flight Routes and Fares 1993-2024.csv")  # folder name + file
df.head()
```

```python
df.drop(['tbl', 'citymarketid_1','citymarketid_2','airportid_1','airportid_2','Geocoded_City1','Geocoded_City2','tbl1apk'], axis=1, inplace=True)
df.head()
```

```python
df.drop(['airport_1', 'airport_2'], axis=1, inplace=True)
df.head()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

#  Fare vs Passengers
axes[0].scatter(df['fare'], df['passengers'])
axes[0].set_xlabel("Fare")
axes[0].set_ylabel("Passengers")
axes[0].set_title("Fare vs Passengers")

# Fare vs Nsmiles
axes[1].scatter(df['fare'], df['nsmiles'])
axes[1].set_xlabel("Fare")
axes[1].set_ylabel("Nsmiles (Distance)")
axes[1].set_title("Fare vs Distance")

# Fare vs Large_ms
axes[2].scatter(df['fare'], df['large_ms'])
axes[2].set_xlabel("Fare")
axes[2].set_ylabel("Large Market Share")
axes[2].set_title("Fare vs Competition")

plt.tight_layout()
plt.show()
```

```python
print(df[['fare', 'passengers', 'nsmiles', 'large_ms']].corr())
```

```python
# âœ… Create Route ID column
df['route_id'] = df['city1'] + "-" + df['city2']

# âœ… Average Competitor Fare column
df['avg_competitor_fare'] = (df['fare_lg'] + df['fare_low']) / 2

# âœ… Approx. Demand Index column
df['demand_index'] = df['passengers'] / df['nsmiles'].replace(0, pd.NA)
df['demand_index'] = df['demand_index'].fillna(0)

df.head()
```

```python
yearly_fare = df.groupby('Year')['fare'].mean().reset_index()
yearly_fare
```

```python
plt.figure(figsize=(8,5))
plt.bar(yearly_fare['Year'], yearly_fare['fare'])
plt.xlabel("Year")
plt.ylabel("Average Fare")
plt.title("Year-wise Average Fare")
plt.show()
```
